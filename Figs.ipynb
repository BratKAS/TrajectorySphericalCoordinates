```python
from sympy import *

init_printing(use_unicode=True)

x = symbols('x')
y1 = x**3
y2 = y1 + 1
y3 = y2.subs(x, x-3)

p = plot(y1, y2, y3, (x, -3, 6), xlabel='$x$', ylabel='$y(x)$')
```

```python
y1 = 1/x
y2 = y1*2
y3 = y2 + 4
y4 = y3.subs(x, x - 1)
y5 = abs(y4)
p = plot(y1, y2, y3, y4, y5, (x, -5, 5), ylim=(-5, 9), xlabel='$x$', ylabel='$y(x)$')
```
